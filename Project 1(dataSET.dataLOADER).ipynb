```python
import torch
import torch.nn as nn
import json
from PIL import Image
import numpy as np
from torch.utils.data import Dataset, DataLoader
import os    
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

```python
class Dataset(Dataset):
    
    def __init__(self, *args): #получаем n массивов данных
        cnt, labels, temp =0, [], []
        self.data = []
        for data in args: #выделяем по одному из них
            self.data.extend(torch.tensor(data).type(torch.float)) # преобразуем каждый в тензор и добавляем к общим данным
            for j in range(len(data)):
                labels.append(cnt) #присовим номера каждому подмассиву для проверки
            cnt += 1
        self.labels = labels
            
    def __getitem__(self, idx):
        return self.data[idx], self.labels[idx]
    
    def __len__(self):
        return len(self.data)
```

```python
data, temp = [],[]
gests = os.listdir('dataset') #884 & 686
for gest in gests:
    data = []
    f = open('dataset/' + str(gest)).readline()
    f = json.loads(f)
    for i in range (len(f)-100):
        data.append(f[i])
    temp.append(data)
    print(len(data))
#     print(data)
data = np.array(temp, dtype=object)
len(data)
```
